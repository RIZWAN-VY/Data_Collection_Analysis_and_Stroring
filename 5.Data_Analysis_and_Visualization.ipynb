# DATA ANALYSIS AND VISUALIZATION 

```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# Load the cleaned Data
df = pd.read_csv(r"C:\Users\rizwa\Rizwan\Projects\Data_Collection_and_Storage\4.Cleaned_British_Airways_Customer_reviews.csv")
print(df)
```

### Total Reviews Collected

```python
reviews_collected = "Total number of Reviews Collected from Skytrax Website :"
count = len(df)
print(reviews_collected,count)
```

### Date Range of Reviews Collected

```python
first_value = df["Date_Review_Posted"].iloc[0]
last_value = df["Date_Review_Posted"].iloc[-1]
print(f"Reviews Collected Between:{first_value} and {last_value}")  # Added "and" for better readability
```

### Average Rating

```python
average_rating = df['Ratings'].mean()
print(f"Average Rating: {average_rating:.2f}")
```

### Sentiment Analysis

```python
sentiment_counts = df['Review_Sentiment'].value_counts()

# Visualize the sentiment counts using a bar chart
plt.figure(figsize=(12, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'gold'])
plt.title('Review Sentiment Counts', fontsize=15,fontweight='bold')
plt.xlabel('Review Sentiment',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Add count values on top of each bar
for i, v in enumerate(sentiment_counts):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=12)
plt.tight_layout()

# Print the sentiment counts (optional)
print(sentiment_counts)

# Show the plot
plt.show()
```

#### Sentiment Analysis of Customer Reviews :
Sentiment Analysis was performed on customer reviews using TextBlob.
Sentiments were categorized into three groups: Positive, Negative, and Neutral.
The sentiment distribution reveals:
* __Positive : 1678__ reviews are positive, indicating a satisfied customer base.  
* __Neutral : 8__ reviews are neutral, suggesting a balanced sentiment.  
* __Negative : 2104__ reviews are negative, signaling areas that might need improvement.

### Recommendation

```python
recommended_counts = df['Recommended'].value_counts()

plt.figure(figsize=(12, 6))
recommended_counts.plot(kind='bar')
plt.title('Recommended Counts', fontsize=15,fontweight='bold')
plt.xlabel('Recommended',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Add count values on top of each bar
for i, v in enumerate(recommended_counts):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
print(f"recommended_counts: {recommended_counts}")
```

## Word Cloud

```python
reviews = ' '.join(df['Review'].tolist() + df['Detailed_Review'].tolist()) # Combine reviews into a single string

stopwords = set(w.lower() for w in STOPWORDS)  # Convert stopwords to lowercase
stopwords.update(['ba', 'flight','british','airways','airline'])  # Add stopwords in lowercase to remove common

wordcloud = WordCloud(width=400, height=300, background_color='white',stopwords=stopwords, max_words=500).generate(reviews)

plt.figure(figsize=(12,6), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
